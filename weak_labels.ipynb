```python
import pandas as pd

csv_path = "/content/drive/MyDrive/meddiag/multimodal_cases.csv"

df = pd.read_csv(csv_path)

print(df.shape)
df.head()
```

```python
SYSTEM_LABELS = [
    "neurological",
    "cardiovascular",
    "respiratory",
    "musculoskeletal",
    "hematological",
    "hepatobiliary",
    "gastrointestinal",
    "genitourinary",
    "renal",
    "lymphatic",
    "immunological",
    "dermatological",
    "ophthalmic",
    "ent",   # ear nose throat
    "endocrine",
    "multisystemic"
]
```

```python
def build_input(row):
    age = str(row.get('age', ''))
    gender = str(row.get('gender', ''))

    return (
        f"Age: {age}. Gender: {gender}. " +   #  structured context
        str(row.get('title', '')) + " " +
        str(row.get('keywords', '')) * 2 + " " +
        str(row.get('abstract', '')) + " " +
        str(row.get('image_captions', ''))
    )
```

```python
df['input_text'] = df.apply(build_input, axis=1)
```

```python
from sentence_transformers import SentenceTransformer
import numpy as np

model = SentenceTransformer('all-MiniLM-L6-v2')
```

```python
system_emb = model.encode(SYSTEM_LABELS)
```

```python
def predict_system(text):
    emb = model.encode([text])[0]
    sims = np.dot(system_emb, emb)
    return SYSTEM_LABELS[np.argmax(sims)]
```

```python
from tqdm import tqdm

batch_size = 64
texts = df['input_text'].tolist()

system_preds = []

for i in tqdm(range(0, len(texts), batch_size)):
    batch = texts[i:i+batch_size]

    emb = model.encode(batch)
    sims = np.dot(emb, system_emb.T)

    preds = [SYSTEM_LABELS[np.argmax(s)] for s in sims]
    system_preds.extend(preds)

df['system_label'] = system_preds
```

```python
def system_rule_boost(text):
    text = str(text).lower()

    #  STRONG organ keywords

    # neurological
    if any(x in text for x in ['brain', 'mening', 'neuro', 'spinal', 'arachnoid']):
        return 'neurological'

    # cardiovascular
    if any(x in text for x in ['heart', 'cardiac', 'aortic']):
        return 'cardiovascular'

    # respiratory
    if any(x in text for x in ['lung', 'pulmonary']):
        return 'respiratory'

    # hepatobiliary
    if any(x in text for x in ['liver', 'hepatic']):
        return 'hepatobiliary'

    #  NEW: spleen → NOT hepatobiliary
    if 'spleen' in text or 'splenic' in text:
        return 'hematological'

    # renal
    if any(x in text for x in ['kidney', 'renal', 'neph']):
        return 'renal'

    # dermatological (important for infections)
    if any(x in text for x in ['skin', 'cutaneous', 'bite', 'wound']):
        return 'dermatological'

    # ophthalmic
    if any(x in text for x in ['eye', 'retina', 'cornea', 'ophthalm']):
        return 'ophthalmic'

    # ENT
    if any(x in text for x in ['nasal', 'sinus', 'oral', 'tooth', 'dental']):
        return 'ent'

    # genitourinary
    if any(x in text for x in ['uterus', 'ovary', 'prostate', 'bladder']):
        return 'genitourinary'

    # musculoskeletal
    if any(x in text for x in ['bone', 'muscle', 'tendon', 'fracture']):
        return 'musculoskeletal'

    # genetic / syndrome
    if 'syndrome' in text:
        return 'genetic'

    # drug / therapy → multisystemic
    if any(x in text for x in ['therapy', 'treatment', 'drug']):
        return 'multisystemic'

    return None
```

```python
def predict_system_final(text):
    rule = system_rule_boost(text)

    if rule:
        return rule

    return predict_system(text)
```

```python
df['system_label'] = df['input_text'].apply(predict_system_final)
```

```python
def predict_system_final(text):
    text = str(text)

    #  Step 1: model prediction
    emb = model.encode([text])[0]
    sims = np.dot(system_emb, emb)

    model_pred = SYSTEM_LABELS[np.argmax(sims)]
    model_conf = np.max(sims)

    #  Step 2: rule prediction
    rule_pred = system_rule_boost(text)

    #  Step 3: combine logic

    # case 1: no rule → trust model
    if rule_pred is None:
        return model_pred

    # case 2: strong model confidence → trust model
    if model_conf > 0.6:
        return model_pred

    # case 3: otherwise use rule
    return rule_pred
```

```python
from tqdm import tqdm

batch_size = 64
texts = df['input_text'].tolist()

final_preds = []

for i in tqdm(range(0, len(texts), batch_size)):
    batch = texts[i:i+batch_size]

    #  model embeddings (batch)
    emb = model.encode(batch)
    sims = np.dot(emb, system_emb.T)

    model_preds = [SYSTEM_LABELS[np.argmax(s)] for s in sims]
    model_confs = [np.max(s) for s in sims]

    #  rule predictions
    rule_preds = [system_rule_boost(t) for t in batch]

    #  combine
    batch_preds = []

    for mp, mc, rp in zip(model_preds, model_confs, rule_preds):

        if rp is None:
            batch_preds.append(mp)

        elif mc > 0.6:
            batch_preds.append(mp)

        else:
            batch_preds.append(rp)

    final_preds.extend(batch_preds)
```

```python
df['system_label'] = final_preds
```

```python
TYPE_LABELS = [
    "tumor",
    "infection",
    "vascular",
    "metabolic",
    "degenerative",
    "trauma",
    "autoimmune"
]
```

```python
def type_rule_boost(text):
    text = str(text).lower()

    # tumor
    if any(x in text for x in [
        'cancer', 'carcinoma', 'tumor', 'sarcoma',
        'neoplasm', 'metastasis', 'adenoma'
    ]):
        return 'tumor'

    # infection
    if any(x in text for x in [
        'infection', 'bacteria', 'viral', 'fungal',
        'sepsis', 'abscess', 'tuberculosis'
    ]):
        return 'infection'

    # vascular
    if any(x in text for x in [
        'stroke', 'aneurysm', 'thrombosis', 'embolism'
    ]):
        return 'vascular'

    # metabolic
    if any(x in text for x in [
        'diabetes', 'metabolic', 'obesity'
    ]):
        return 'metabolic'

    # trauma
    if any(x in text for x in [
        'injury', 'fracture', 'trauma'
    ]):
        return 'trauma'

    # autoimmune
    if any(x in text for x in [
        'autoimmune', 'lupus', 'arthritis'
    ]):
        return 'autoimmune'

    # degenerative
    if any(x in text for x in [
        'degenerative', 'parkinson', 'alzheimer'
    ]):
        return 'degenerative'

    return None
```

```python
type_emb = model.encode(TYPE_LABELS)
```

```python
def predict_type(text):
    emb = model.encode([text])[0]
    sims = np.dot(type_emb, emb)
    return TYPE_LABELS[np.argmax(sims)], np.max(sims)
```

```python
def predict_type_final(text):
    text = str(text)

    # rule
    rule = type_rule_boost(text)

    # model
    model_pred, model_conf = predict_type(text)

    # combine
    if rule is None:
        return model_pred

    if model_conf > 0.7:
        return model_pred

    return rule
```

```python
from tqdm import tqdm

batch_size = 64
texts = df['input_text'].tolist()

type_preds = []

for i in tqdm(range(0, len(texts), batch_size)):
    batch = texts[i:i+batch_size]

    # model
    emb = model.encode(batch)
    sims = np.dot(emb, type_emb.T)

    model_preds = [TYPE_LABELS[np.argmax(s)] for s in sims]
    model_confs = [np.max(s) for s in sims]

    # rules
    rule_preds = [type_rule_boost(t) for t in batch]

    batch_preds = []

    for mp, mc, rp in zip(model_preds, model_confs, rule_preds):

        if rp is None:
            batch_preds.append(mp)

        elif mc > 0.7:
            batch_preds.append(mp)

        else:
            batch_preds.append(rp)

    type_preds.extend(batch_preds)
```

```python
df['type_label'] = type_preds
```

```python
import pandas as pd

df = pd.read_csv("/content/final_labeled_dataset.csv")

df = df.dropna()
df = df.reset_index(drop=True)

print(df.shape)
print(df.head())
```

```python
from sklearn.preprocessing import LabelEncoder

le_system = LabelEncoder()
le_type = LabelEncoder()

df['system_id'] = le_system.fit_transform(df['system_label'])
df['type_id'] = le_type.fit_transform(df['type_label'])
```

```python
from sklearn.model_selection import train_test_split

train_df, val_df = train_test_split(
    df, test_size=0.1, random_state=42, stratify=df['system_id']
)
```

```python
from transformers import AutoTokenizer

tokenizer = AutoTokenizer.from_pretrained("distilbert-base-uncased")
```

```python
import torch
from torch.utils.data import Dataset

class MedicalDataset(Dataset):
    def __init__(self, df, tokenizer):
        self.texts = df['input_text'].tolist()
        self.system = df['system_id'].tolist()
        self.type = df['type_id'].tolist()
        self.tokenizer = tokenizer

    def __len__(self):
        return len(self.texts)

    def __getitem__(self, idx):
        encoding = self.tokenizer(
            self.texts[idx],
            padding='max_length',
            truncation=True,
            max_length=512,
            return_tensors='pt'
        )

        return {
            'input_ids': encoding['input_ids'].squeeze(),
            'attention_mask': encoding['attention_mask'].squeeze(),
            'system_label': torch.tensor(self.system[idx]),
            'type_label': torch.tensor(self.type[idx])
        }
```

```python
from torch.utils.data import DataLoader

train_dataset = MedicalDataset(train_df, tokenizer)
val_dataset = MedicalDataset(val_df, tokenizer)

train_loader = DataLoader(train_dataset, batch_size=8, shuffle=True)
val_loader = DataLoader(val_dataset, batch_size=8)
```

```python
from transformers import AutoModel
import torch.nn as nn

class MultiTaskModel(nn.Module):
    def __init__(self, num_system, num_type):
        super().__init__()

        self.encoder = AutoModel.from_pretrained("distilbert-base-uncased")
        hidden = self.encoder.config.hidden_size

        self.dropout = nn.Dropout(0.3)

        self.system_head = nn.Linear(hidden, num_system)
        self.type_head = nn.Linear(hidden, num_type)

    def forward(self, input_ids, attention_mask):
        outputs = self.encoder(input_ids=input_ids, attention_mask=attention_mask)

        cls = outputs.last_hidden_state[:, 0]
        cls = self.dropout(cls)

        system_logits = self.system_head(cls)
        type_logits = self.type_head(cls)

        return system_logits, type_logits
```

```python
device = torch.device("cuda" if torch.cuda.is_available() else "cpu")

model = MultiTaskModel(
    num_system=len(le_system.classes_),
    num_type=len(le_type.classes_)
).to(device)

optimizer = torch.optim.AdamW(model.parameters(), lr=2e-5)

loss_fn = nn.CrossEntropyLoss()
```

```python
from tqdm import tqdm

epochs = 3

for epoch in range(epochs):
    model.train()
    total_loss = 0

    for batch in tqdm(train_loader):
        optimizer.zero_grad()

        input_ids = batch['input_ids'].to(device)
        attention_mask = batch['attention_mask'].to(device)
        sys_labels = batch['system_label'].to(device)
        type_labels = batch['type_label'].to(device)

        sys_logits, type_logits = model(input_ids, attention_mask)

        loss_sys = loss_fn(sys_logits, sys_labels)
        loss_type = loss_fn(type_logits, type_labels)

        loss = loss_sys + loss_type

        loss.backward()
        optimizer.step()

        total_loss += loss.item()

    print(f"Epoch {epoch+1}, Loss: {total_loss}")
```

```python
from sklearn.metrics import f1_score

model.eval()

sys_preds, sys_true = [], []
type_preds, type_true = [], []

with torch.no_grad():
    for batch in val_loader:

        input_ids = batch['input_ids'].to(device)
        attention_mask = batch['attention_mask'].to(device)

        sys_logits, type_logits = model(input_ids, attention_mask)

        sys_pred = torch.argmax(sys_logits, dim=1).cpu().numpy()
        type_pred = torch.argmax(type_logits, dim=1).cpu().numpy()

        sys_preds.extend(sys_pred)
        type_preds.extend(type_pred)

        sys_true.extend(batch['system_label'].numpy())
        type_true.extend(batch['type_label'].numpy())
```

```python
print("System F1:", f1_score(sys_true, sys_preds, average='weighted'))
print("Type F1:", f1_score(type_true, type_preds, average='weighted'))
```
